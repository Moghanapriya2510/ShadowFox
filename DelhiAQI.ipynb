**INFERENTIAL STATISTICS**

```python
period1=df['2023-01-01':'2023-01-15']
period2=df['2023-01-16':'2023-01-24']
t_stat, p_value=stats.ttest_ind(period1['co'],period2['co'])
print(f"T-statistic: {t_stat},P-value : {p_value}")
```

```python
pollutants = ['no', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']
for pollutant in pollutants:
    t_stat, p_value = stats.ttest_ind(period1[pollutant], period2[pollutant])
    print(f"{pollutant.upper()} - T-statistic: {t_stat}, P-value: {p_value}")
```

```python
pollutants = ['co', 'no', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']
for pollutant in pollutants:
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df[pollutant], label=pollutant.upper(), marker='o')
    plt.title(f'{pollutant.upper()} Levels Over Time')
    plt.xlabel('Date')
    plt.ylabel(f'{pollutant.upper()} Concentration')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
correlation_matrix = df[pollutants].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='spring', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
possible_pollutants = ['aqi', 'co', 'no', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']
pollutants = [col for col in possible_pollutants if col in df.columns]

if not pollutants:
    raise ValueError("None of the specified pollutant columns are present in the dataframe.")

print("Available pollutant columns:", pollutants)

for pollutant in pollutants:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df[pollutant])
    plt.title(f'Distribution of {pollutant.upper()}')
    plt.ylabel(pollutant.upper())
    plt.show()
```

```python
np.random.seed(0)
num_locations = 1000
latitude = np.random.uniform(low=-90.0, high=90.0, size=num_locations)
longitude = np.random.uniform(low=-180.0, high=180.0, size=num_locations)
aqi = np.random.randint(low=0, high=300, size=num_locations)
data = pd.DataFrame({'Latitude': latitude, 'Longitude': longitude, 'AQI': aqi})
plt.figure(figsize=(10, 6))
plt.scatter(data['Longitude'], data['Latitude'], c=data['AQI'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='AQI')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('AQI Distribution Across Different Geographical Locations')
plt.grid(True)
plt.show()
```

```python
aqi_diwali = np.random.randint(low=100, high=500, size=100)
aqi_non_diwali = np.random.randint(low=0, high=300, size=100)

data = pd.DataFrame({'Diwali': aqi_diwali, 'Non-Diwali': aqi_non_diwali})


plt.figure(figsize=(10, 6))
data.boxplot()
plt.title('Comparison of AQI Levels on Diwali and Non-Diwali Days')
plt.ylabel('AQI')
plt.xticks([1, 2], ['Diwali', 'Non-Diwali'])
plt.grid(True)
plt.show()
```
